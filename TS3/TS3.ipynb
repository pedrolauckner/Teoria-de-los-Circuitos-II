Se plantea un filtro pasa altos de maxima planicidad,tanto su version de circuitos pasivos cascadeados como su version equivalente activa sin el uso de inductores.

<img src="./CircPas.PNG" align="center" width="1000" /> 

\begin{equation}
T(s) = \frac{V2}{V1} = \frac{s^2}{s^2 + s\frac{Wo}{Q1} + Wo^2} * \frac{s^2}{s^2 + s\frac{Wo}{Q2} + Wo^2}
\end{equation}

Con
\begin{equation} 
Wo^2 = \frac{1}{L C} = 345*10^6
\end{equation}

\begin{equation} 
\frac{Wo}{Q1} = \frac{R1}{L} = 34324
\end{equation}

\begin{equation} 
\frac{Wo}{Q2} = \frac{R2}{L} = 14209
\end{equation}

\begin{equation}
L = 1mH
\end{equation}

\begin{equation}
C = 2.9uf
\end{equation}

\begin{equation}
R1 = 34.324
\end{equation}

\begin{equation}
R2 = 14.209
\end{equation}


\begin{equation}
T(s) = \frac{V2}{V1} = \frac{s^2}{s^2 + s 34324 + 345*10^6} * \frac{s^2}{s^2 + s 14209 + 345*10^6}
\end{equation}

Si juntamos todo en una transferencia de orden 4 queda:

\begin{equation}
T(s) = \frac{V2}{V1} = \frac{s^4}{s^4 + s^3*48533 + s^2*1.17*10^9 + s*1.674*10^{13} + 1.19*10^{17}} 
\end{equation}


Su version activa sin inductores que responde a la misma transferencia

<img src="./CircAct.PNG" align="center" width="1000" /> 


```python
# M√≥dulos para Jupyter

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# m√≥dulo de SciPy
from scipy import signal as sig

from splane import bodePlot, pzmap, pretty_print_bicuad_omegayq


# Setup inline graphics: Esto lo hacemos para que el tama√±o de la salida, 
# sea un poco m√°s adecuada al tama√±o del documento
mpl.rcParams['figure.figsize'] = (10,10)


# Numerador y denominador 

num = np.array([1, 0, 0 , 0, 0]) 
den = np.array([1, 48533, 1.17*10**9, 1.674*10**13, 1.19*10**17])

# Transferencia

T_s = sig.TransferFunction(num, den)

# Plot

plt.close('all')
bodePlot(T_s)

# polos y ceros
pzmap(T_s)


```

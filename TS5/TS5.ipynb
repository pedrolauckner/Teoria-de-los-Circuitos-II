Ejercicio 1

Se busca implimentar un filtro pasatodo de primer orden que para w=1 desfase 15° respecto w=0

Para esto se propone el circuito pasivo lattice de primer orden

<img src="./Circ1.PNG" align="center" width="600" /> 

\begin{equation}
T(s) = \frac{V2}{V1} = \frac{s - \frac{1}{C R}}{s + \frac{1}{C R}}
\end{equation}

Donde la fase es:

\begin{equation}
θ(w) = arctg(\frac{w}{\frac{-1}{C R}}) - arctg(\frac{w}{\frac{1}{C R}})
\end{equation}

Por propiedad arctg(x) = - arctg(-x) por lo que:

\begin{equation}
θ(w) = -2 arctg(W C R)
\end{equation}

Para θ(1) = θ(0) - 15°

\begin{equation}
R = 1
\end{equation}

\begin{equation}
C = 131.65 mF
\end{equation}


```python
# Módulos para Jupyter

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# módulo de SciPy
from scipy import signal as sig

from splane import bodePlot, pzmap, pretty_print_bicuad_omegayq


# Setup inline graphics: Esto lo hacemos para que el tamaño de la salida, 
# sea un poco más adecuada al tamaño del documento
mpl.rcParams['figure.figsize'] = (10,10)


# Numerador y denominador 
# Antes de desnormalizar

num = np.array([1, -7.5959]) 
den = np.array([1, 7.5959])

# Transferencia

T_s = sig.TransferFunction(num, den)

# Plot

plt.close('all')
bodePlot(T_s, [-2,3],[10,-10])

# polos y ceros
pzmap(T_s)


```

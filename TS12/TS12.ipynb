```python
# M√≥dulos para Jupyter

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# m√≥dulo de SciPy
from scipy import signal as sig

from splane import bodePlot, pzmap, pretty_print_bicuad_omegayq


# Setup inline graphics: Esto lo hacemos para que el tama√±o de la salida, 
# sea un poco m√°s adecuada al tama√±o del documento
mpl.rcParams['figure.figsize'] = (10,10)


# Numerador y denominador Y11

num = np.array([3, 0, 7, 0]) 
den = np.array([1, 0, 7, 0, 10])

# Transferencia

T_s = sig.TransferFunction(num, den)

# Plot

plt.close('all')
bodePlot(T_s, [-1,1],[50,-40])

# polos y ceros
pzmap(T_s)

```

```python
#########################################################
# Numerador y denominador Y21

num = np.array([1, 0, 1, 0]) 
den = np.array([1, 0, 7, 0, 10])

# Transferencia

T_s = sig.TransferFunction(num, den)

# Plot

plt.close('all')
bodePlot(T_s, [-2,2],[50,-60])

# polos y ceros
pzmap(T_s)
```

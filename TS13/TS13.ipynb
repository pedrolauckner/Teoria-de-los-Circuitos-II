Transferencia cargada de un cuadripolo RC

\begin{equation}
\frac{-I2}{I1} = H * \frac{s^2 + 5s + 4}{s^2 + 8s + 12}
\end{equation}

\begin{equation}
Z21 = 6H
\end{equation}


<img src="./Circ.PNG" align="center" width="1000" /> 


Simulaciones en LTSpice:

<img src="./SIM.PNG" align="center" width="1000" /> 



```python
# M√≥dulos para Jupyter

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# m√≥dulo de SciPy
from scipy import signal as sig

from splane import bodePlot, pzmap, pretty_print_bicuad_omegayq


# Setup inline graphics: Esto lo hacemos para que el tama√±o de la salida, 
# sea un poco m√°s adecuada al tama√±o del documento
mpl.rcParams['figure.figsize'] = (10,10)


num = np.array([1, 5, 4]) 
den = np.array([1, 8, 12])

# Transferencia

T_s = sig.TransferFunction(num, den)

# Plot

plt.close('all')
bodePlot(T_s, [-2,2],[5,-15])

# polos y ceros
pzmap(T_s)

```

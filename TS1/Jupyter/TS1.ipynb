Se plantea el siguiente circuito pasa todo de primer orden y se obtiene su transferencia $H(s)$ 

<img src="./Circuito.PNG" align="center" width="500" /> 

\begin{equation}
H(s) = \frac{V2}{V1} = \frac{s-\frac{R2}{C R3 R1}}{s+\frac{1}{C R3}}.
\end{equation}

Ademas podemos sacar su modulo

\begin{equation}
|H(w)| = \frac{\sqrt{w^2+(\frac{R2}{C R3 R1})^2}}{\sqrt{w^2+(\frac{1}{C R3})^2}}.
\end{equation}

Si R2=R1, el modulo es 1 para todas las frecuencias

Y la fase

\begin{equation}
φ(w) = arctg(\frac{w C R3 R1}{R2}) - arctg(w C R3 R1)
\end{equation}


Luego se puede normalizar en frecuencia

Planteamos
\begin{equation}
Wa = \frac{1}{C R3}
\end{equation}

lo que cambia

\begin{equation}
C' = C * Wa = \frac{1}{C R3}
\end{equation}


\begin{equation}
H(s) = \frac{s-\frac{R2}{R1}}{s+1}.
\end{equation}

Luego en LTSpice analizamos el circuito normalizado
Considerando una normalizacion en frecuencia

\begin{equation}
Wa = \frac{1}{C R3}
\end{equation}

Y una normalizacion en impedancia
\begin{equation}
R3 = 1
\end{equation}

Por lo que:
\begin{equation}
R1' = \frac{R1}{R3} \\
R2' = \frac{R2}{R3} \\
R3' = 1 \\
C' = C*\frac{1}{C R3} = 1
\end{equation}
